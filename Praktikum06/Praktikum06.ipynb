**Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Dataset**

```python
# Dataset
stores = ['Store A', 'Store B', 'Store C']
male = [150, 200, 180]
female = [150, 300, 200]

product_a = [200, 250, 300]
product_b = [150, 300, 200]

# Data quarter
q1_male = [150, 180, 160]
q1_female = [140, 200, 150]
q2_male = [170, 190, 175]
q2_female = [130, 210, 160]
q3_male = [160, 185, 170]
q3_female = [135, 195, 165]
q4_male = [155, 175, 160]
q4_female = [125,205, 150]
```

**Basic Stacked Vertikal Bar Chart untuk Gender**

```python
plt.figure(figsize=(7,5))
x = np.arange(len(stores))

plt.bar(x, male, label='Male', color='blue')
plt.bar(x, female, bottom=male, label='Female', color='pink')

plt.title('Population by Gender and Store')
plt.xlabel('Store')
plt.ylabel('Population')
plt.xticks(x, stores)
plt.legend()
plt.show()
```

**Basic Stacked Vertikal Bar Chart untuk Product**

```python
plt.figure(figsize=(7,5))
x = np.arange(len(stores))

plt.bar(x, product_a, label='Product A', color='blue')
plt.bar(x, product_b, bottom=product_a, label='Product B', color='green')

plt.title('Sales Transaktion by Store')
plt.xlabel('Store')
plt.ylabel('Sales')
plt.xticks(x, stores)
plt.legend()
plt.show()
```

**Kustomisasi Stacked Vertikal Bar Chart**

```python
plt.figure(figsize=(7,5))
x = np.arange(len(stores))

plt.bar(x, product_a, label='Product A', color='blue')
plt.bar(x, product_b, bottom=product_a, label='Product B', color='green')

# Tambah angka (sama seperti Streamlit)
for i in range(len(x)):
    plt.text(x[i], product_a[i]/2, str(product_a[i]), ha='center', color='white')
    plt.text(x[i], product_a[i] + product_b[i]/2, str(product_b[i]), ha='center', color='white')

plt.title('Kustomisasi Stacked Vertikal Bar Chart')
plt.xlabel('Store')
plt.ylabel('Sales')
plt.xticks(x, stores)
plt.legend()
plt.show()
```

**Multiple Stacked Vertikal Bar Chart**

```python
width = 0.2
x = np.arange(len(stores))

plt.bar(x-width * 1.5, q1_male, label='Q1 Male', color='lightblue', width=width)
plt.bar(x-width * 1.5, q1_female, bottom=q1_male, label='Q1 Female', color='pink', width=width)

plt.bar(x-width/2, q2_male, label='Q2 Male', color='blue', width=width)
plt.bar(x-width/2, q2_female, bottom=q2_male, label='Q2 Female', color='red', width=width)

plt.bar(x+width/2, q3_male, label='Q3 Male', color='green', width=width)
plt.bar(x+width/2, q3_female, bottom=q3_male, label='Q3 Female', color='yellow', width=width)

plt.bar(x+width * 1.5, q4_male, label='Q4 Male', color='purple', width=width)
plt.bar(x+width * 1.5, q4_female, bottom=q4_male, label='Q4 Female', color='orange', width=width)

plt.title('Population by Gender and Store (Mutltiple Quarters)')
plt.ylabel('Stores')
plt.xlabel('Population')
plt.xticks(x, stores)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
