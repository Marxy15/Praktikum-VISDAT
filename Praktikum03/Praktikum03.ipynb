**Praktikum 3 : Matlotlib - Line Plot**
** Kelompok :**

```python
import matplotlib.pyplot as plt
```

```python
# Buat data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

```python
# Line plot 1 garis, style grafik default
plt.plot(months,product_A_sales, label='Product A')
plt.title('Penjualan A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show
```

```python
#Costumisasi Line Plot (Line 1 produk A)
plt.plot(months,product_A_sales, label='Product A',linestyle="-", marker="o",color='blue')
#Costumisasi Line Plot (Line 1 produk A)
plt.plot(months,product_B_sales, label='Product B',linestyle="-", marker="x",color='red')

plt.title('Penjualan A dan B per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show
```

```python
# Data sample tambahan
product_C_sales = [18,22,15,28,32,38,42,45,48,52,56,60]
product_D_sales = [23,54,63,68,45,33,23,34,54,20,25,30]

#Costumisasi Line Plot (Line 1 produk A)
plt.plot(months,product_A_sales, label='Product A',linestyle="--",color='blue')
#Costumisasi Line Plot (Line 1 produk A)
plt.plot(months,product_B_sales, label='Product B',linestyle="-",color='red')
#Costumisasi Line Plot (Line 1 produk A)
plt.plot(months,product_C_sales, label='Product C',linestyle=":",color='green')
#Costumisasi Line Plot (Line 1 produk A)
plt.plot(months,product_D_sales, label='Product D',linestyle="-.",color='purple')

plt.title('Penjualan Product per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show
```

```python
# subplot, 2 garis 1 kolom, panjang 10 inch, tinggi 8 inch
fig, axs = plt.subplots(2, 1, figsize=(10,8))

axs[0].plot(months,product_A_sales, label='Product A', marker="o",color='blue')
axs[0].set_title('Penjualan Product A per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Penjualan')
axs[0].legend()
axs[0].grid(True)

axs[1].plot(months,product_A_sales, label='Product A', marker="o",color='blue')
axs[1].set_title('Penjualan Product A per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
